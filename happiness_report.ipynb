```python
import pandas as pd
import seaborn as sns
import plotly.graph_objs as go 
from plotly.offline import init_notebook_mode,iplot,plot
init_notebook_mode(connected=True) 
```

### Read Data

```python
df2015 = pd.read_csv('./data/happiness_report/2015.csv')
df2016 = pd.read_csv('./data/happiness_report/2016.csv')
df2017 = pd.read_csv('./data/happiness_report/2017.csv')
```

### Info

```python
df2015.info()
df2016.info()
df2017.info()
```

### World Map with Happiness Rank

```python
data = dict(
        type = 'choropleth',
        colorscale = 'Viridis',
        reversescale = True,
        locations = df2015['Country'],
        locationmode = "country names",
        z = df2015['Happiness Rank'],
        text = df2015['Country'],
        colorbar = {'title' : 'Happiness Rank'},
      ) 

layout = dict(title = '2015 Happiness Report',
                geo = dict(showframe = False,projection = {'type':'mercator'})
             )
```

```python
choromap = go.Figure(data = [data],layout = layout)
iplot(choromap,validate=False)
```
